```python
# 1º Importar base de dados
import pandas as pd
import numpy

dados = pd.read_csv("/content/clientes.csv")
display(dados)
```

```python
print(dados.info())
```

```python
# Tratando os dados para a IA

from sklearn.preprocessing import LabelEncoder
# serve para converter variaveis que possuem relação de ordem

# > aplicando a ferramenta LabelEncoder ana coluna profissão
codificador = LabelEncoder() #codificador é a ferramenta que vai transformar nosso texto em número
dados["profissao"] = codificador.fit_transform(dados["profissao"]) #quem faz a transformação é o método fit_transform

dados["mix_credito"] = codificador.fit_transform(dados["mix_credito"])

dados["comportamento_pagamento"] = codificador.fit_transform(dados["comportamento_pagamento"])
print(dados.info())

# > Com isso, tudo que estava com objet no Dtype especifico no arquivo foi transformado para int64
```

```python
# * Separar  as informações em dados de treino e dados de teste
y = dados["score_credito"]
# * todas as colunas que a gente vai usar para fazer a previsão do score credito
x = dados.drop(["score_credito","id_cliente"], axis=1)

# > OBS: Para que uma IA entenda o que deve ser feito , ela primeira estuda varios dados que são passados anteriomente e dai aplica seus conhecimentos em cima de dados
# para isso acontecer, ele seleciona dados de treino (70%) e dados de teste(30%) , mas que pode ser definido...que é dado por essa ferramenta:

from sklearn.model_selection import train_test_split

# e assim, deve seguir essa ordem:

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y)
```

**Criar um Modelo**
> árvore de decisão = modelo baseado em perguntas

 > KNN - nearest neighbors

 * importa
 * cria
 * treina



```python
# Importando as 2 IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

#criando:

modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

#treinar:

modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)

#cada um aprende da sua forma essa é a diferença entre elas, a forma como elas treinam
```

```python
# Após treinar , feito no código anterior
# Ele receberá dados de teste para avaliar qual IA é a melhor escolha para o projeto
# Escolhendo o melhor modelo

from sklearn.metrics import accuracy_score

previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# Comparando
print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_knn))

#vai avaliar qual x_teste se aproxima mais do gabarito que é o y_teste
```

> O melhor modelo é o Árvore de decisão com base em testes
* modelo_arvoredecisao = 83% acurácia
